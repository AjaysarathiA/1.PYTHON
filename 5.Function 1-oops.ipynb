```python
def oddeven():
    num=int(input("ENTER THE NUMBER: "))
    if(num%2==0):
        print("Even number")
        message="Even number"
    else:
        print("Odd number")
        message="Odd number"
    return message   
```

```python
message=oddeven()
```

```python
if(message=="Even number"):
 print("The given number is even")
else:
 print("The given number is odd")
```

```python
def BMI():
    BMI=int(input("Enter you BMI: "))
    if(BMI<19.9):
     print("under weight")
     message="under weight"
    elif(BMI<29.9):
      print("normal")
      message="normal"
    elif(BMI<39.9):
      print("over weight")
      message="over weight"
    else:
      print("very overweight")
      message="very overweight"
    return message
```
