```python
def agecal():
    if(age<18):
        print("CHILD")
        category="CHILD"
    elif(age<35):
        print("ADULT")
        category="ADULT"
    elif(age<50):
        print("CITIZEN")
        category="CITIZEN"
    else:
        print("SENIOR CITIZEN")
        category="SENIOR CITIZEN"
    return category
```
